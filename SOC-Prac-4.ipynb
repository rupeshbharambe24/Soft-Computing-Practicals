```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
#Compute the sigmoid of x.
    f(x)=1/ 1+e^-x
#Parameters:
x (numpy array): Input array of values.
Returns:
numpy array: Sigmoid of the input array.

Output: x has large +ve values, output approaches 1
x has large â€“ve values, output approaches 0
"""
    
    return 1 / (1 + np.exp(-x))

if __name__ == '__main__':
# Array of values
    x_values = np.array([-5, -1, 0, 1, 5])

    y_values = sigmoid(x_values)

    print('Input values:', x_values)
    print('Sigmoid values:', y_values)

x_values_dense = np.linspace(-10, 10, 100)
y_values_dense = sigmoid(x_values_dense)

plt.plot(x_values_dense, y_values_dense, label='Sigmoid function')
plt.scatter(x_values, y_values, color='red', label='Sample points')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    '''
Compute the hyperbolic tangent of x.

Parameters:

x (float or numpy array): Input value or array of values.

Returns:
float or numpy array: Hyperbolic tangent of the input value or array.

Formula : - f(x) = (e^x - e^-x) / (e^x + e^-x).
'''
    return np.tanh(x)

if __name__ == '__main__':
# Single value
    x_single = 0.5
print(f'tanh({x_single}) = {tanh(x_single)}')

# Array of values
x_values = np.array([-5, -1, 0, 1, 5])
y_values = tanh(x_values)

print('Input values:', x_values)
print('tanh values:', y_values)

x_values_dense = np.linspace(-10, 10, 100)
y_values_dense = tanh(x_values_dense)

plt.plot(x_values_dense, y_values_dense, label='tanh function')
plt.scatter(x_values, y_values, color='red', label='Sample points')
plt.title('Hyperbolic Tangent Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.legend()
plt.grid(True)
plt.show()
```
