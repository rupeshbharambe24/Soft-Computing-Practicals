```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Generate a range of input values
x = np.linspace(-10, 10, 400)

# Compute the ReLU output
y = relu(x)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the input vs. output of the ReLU function
plt.plot(x, y, label='ReLU', color='r')

# Add labels and title
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ReLU Activation Function')

plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Leaky ReLU activation function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Generate a range of input values
x = np.linspace(-10, 10, 400)

# Compute the Leaky ReLU output
alpha = 0.1
y = leaky_relu(x, alpha)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the input vs. output of the Leaky ReLU function
plt.plot(x, y, label=f'Leaky ReLU (alpha={alpha})', color='b')

# Add labels and title
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Leaky ReLU Activation Function')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
